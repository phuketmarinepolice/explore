# Import Modules

```
import requests
import lxml.etree
```

# Scrape

```
url = 'https://th.wikipedia.org/wiki/รายชื่อเทศบาลตำบลในประเทศไทย'
resp = requests.get(url)

print(resp)
```

```
print(resp.text[:5_000])
```

# Extract Tambons

```
content = resp.content
tree = lxml.etree.fromstring(content, parser=lxml.etree.HTMLParser())

print(tree)
```

```
xpath = '//*[@id="mw-content-text"]/div/table[position() >= 3]/tbody/tr/td[2 <= position() and position() <= 3]//a//text()'
tambon_list = tree.xpath(xpath)

print('got:', len(tambon_list), 'tambons')
print(tambon_list)
```

# **แถม** scrape + extract แบบทั้งตาราง ด้วย pandas
ให้เอาไว้เผื่อใครต้องใช้ครับ แต่จะอธิบายแบบละเอียด ๆ อีกที ในบทความถัด ๆ ไป  
ถ้าใครสนใจรบกวนคอมเม้นบอกในเพจด้วยนะครับ

```
import pandas as pd
```

```
dfs = pd.read_html(content, encoding='utf8')
```

```
columns = ['อำเภอ', 'ลำดับที่', 'ชื่อเทศบาล', 'ครอบคลุมตำบล (เต็มพื้นที่)', 'ครอบคลุมตำบล (บางส่วน)', 'ปีที่ได้รับการจัดตั้ง', 'ชื่อหรือฐานะก่อนหน้า']
```

```
tmp = []
for t in dfs:
    if t.shape[1] != 7:
        continue
    t.columns = columns
    tmp.append(t[columns])

table = pd.concat(tmp).dropna(subset=['ชื่อเทศบาล']).reset_index(drop=True)
```
